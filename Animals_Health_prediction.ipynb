```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
am=pd.read_csv('/content/animal health condition.csv')
```

```python
am
```

```python
am.shape
```

```python
am.describe()
```

```python
am.info()
```

```python
le=LabelEncoder()
```

```python
am['AnimalName']=le.fit_transform(am['AnimalName'])
```

```python
am['symptoms1']=le.fit_transform(am['symptoms1'])
```

```python
am['symptoms2']=le.fit_transform(am['symptoms2'])
```

```python
am['symptoms3']=le.fit_transform(am['symptoms3'])
```

```python
am['symptoms4']=le.fit_transform(am['symptoms4'])
```

```python
am['symptoms5']=le.fit_transform(am['symptoms5'])
```

```python
am['Dangerous']=le.fit_transform(am['Dangerous'])
```

```python
am
```

```python
X=am.drop(columns=['Dangerous'],axis=1)
Y=am['Dangerous']
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=42)
```

```python
model=LogisticRegression()
```

```python
model.fit(X_train,Y_train)
```

```python
X_train_prediction=model.predict(X_train)
traning_data_accuracy=accuracy_score(X_train_prediction,Y_train)
```

```python
print('Accuracy on traning data:',traning_data_accuracy)
```

```python
X_test_prediction=model.predict(X_test)
test_data_accuracy=accuracy_score(X_test_prediction,Y_test)
```

```python
print('accuracy on test data:',test_data_accuracy)
```
